```python
#Romberg Integraion Calculator

import numpy as np

#Criteria for the Estimation and Integration
true_val = 29/6
e_stop = 0.005
e_t = 1.00
A = 1
B = 2

#function of which the integral will be estimated
def func(x):
    return (x + (1/x))**2

#Calculates an estimate of the integral of the function provided and input params:
#n - number of sections in the estimate, b - upper limit of integration, a - lower limit of itegration
def Trapezoid(n, b, a):
    for k in range(1, n+1):
        #print("-----------------", k, "segments--------------")
        I = func(a)
        #print("f(", a, ") =", func(a))

        for i in range(1, k):
            h = (b-a)/k
            I += 2*(func(a+i*h))
            #print("f(", a+i*h, ") =", func(a+i*h))

        I += func(b)
        #print("f(", b, ") =", func(b))

        I *= ((b-a)/(2*k))
        #print("I:", I)
        
    #print("Error:", 100*(abs(e_t - I)/e_t), "%")
    return I

#Calculates the percent error between two values
def error(approx):
    return (100*(abs(true_val-approx)/true_val))

#Romberg Integration requires calculating a tree of values until convergence that keeps growing
#until the stopping criteria of error has been met

trap_iterator = 1
while(e_t > e_stop):
    print("---------------", trap_iterator, "---------------")
    
    I = [None] * trap_iterator
    
    for n in range(0, trap_iterator):
        num_divisions = 2**n
        I[n] = Trapezoid(num_divisions, B, A)
    print(I)
    
    for j in range(0, (len(I)-1)):
        temp_I = [None] * int(len(I) - 1)
        temp_I = I[0:int(len(I)-1)]
        
        for k in range(0, len(temp_I)):
            temp_I[k] = I[k+1] + (1/3)*(I[k+1] - I[k])
        
        I = temp_I
        print(I)
    
    e_t = error(I[0])
    print(e_t)
    trap_iterator +=1
    
```
