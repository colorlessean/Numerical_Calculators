```python
#Gauss-Seidel Solver

#matrix setup of the linear equations
A = [[27,0,5],[0,35,5],[1,1,-1]]
B = [0,-200,0]

#error variables set up
error_required = 0.01
ea_1 = 1.00
ea_2 = 1.00
ea_3 = 1.00

#sets up the guesses and then acts as the storage for the answers
x = [0,0,0]

#iteration variable to keep track of iterations taken
i = 0

while ea_1 > error_required or ea_2 > error_required or ea_3 > error_required:
    #saves program if non-convergent solution
    if(i > 100):
        break
    #iterates iterator
    i+=1
    #stores old values for comparison
    old_x1 = x[0]
    old_x2 = x[1]
    old_x3 = x[2]
    #calculates new values for error calculations
    x[0] = (B[0] - A[0][1]*x[1] - A[0][2]*x[2])/A[0][0]
    x[1] = (B[1] - A[1][0]*x[0] - A[1][2]*x[2])/A[1][1]
    x[2] = (B[2] - A[2][0]*x[0] - A[2][1]*x[1])/A[2][2]
    #calulates error for next iteration
    ea_1 = abs(((x[0] - old_x1))*100)
    ea_2 = abs(((x[1] - old_x2))*100)
    ea_3 = abs(((x[2] - old_x3))*100)
    
    #prints out info for user to record
    print('---------------------------')
    print('iteration: ', i)
    print('x1: ', x[0])
    print('ea1: ', ea_1)
    print('x2: ', x[1])
    print('ea2: ', ea_2)
    print('x3: ',x[2])
    print('ea3: ', ea_3)
```

```python
#nxn Gauss-Seidel Calculator

#matrix setup of the linear equations
A = [[4,-1,-1,0],[-1,4,0,-1],[-1,0,4,-1],[0,-1,-1,4]]
B = [275,225,75,25]

#sets up the guesses and then acts as the storage for the answers
x = [0,0,0,0]

#sets up the error storage
E = [1.0,1.0,1.0,1.0]
#error tolerance
error_required = 0.01

#iterator to break infinite loop
iterator = 0

while(max(E) > error_required and iterator < 100):
    iterator+=1
    print('----------------------------')
    print('iteration: ', iterator)
    x_old = x[:]
    for i in range(0, len(A)):
        x[i] = B[i]
        for j in range(0, len(A[i])):
            if(j != i):
                x[i] -= A[i][j]*x[j]
        x[i] /= A[i][i]
        print('x_', i , ': ', x[i])
        
        if x[i] != 0:
            E[i] = (abs((x[i] - x_old[i])/x[i]))
            print('Ex_', i , ': ', E[i])
        else:
            print('Ex_', i , ': NaN')
            
    
```
