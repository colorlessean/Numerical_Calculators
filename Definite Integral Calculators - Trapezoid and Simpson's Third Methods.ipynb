```python
#Trapezoid Rule Integration Calculator

import numpy as np

def func(x):
    return (51.69375)*np.tanh(0.18977*x)

#limits of integral
b = 10 
a = 0

#n is number of sections, x is the value to calculate for each segment
n = 11

#true value
e_t = 334

for k in range(1, n+1):
    print("-----------------", k, "segments--------------")
    I = func(a)
    print("f(", a, ") =", func(a))
    
    for i in range(1, k):
        h = (b-a)/k
        I += 2*(func(a+i*h))
        print("f(", a+i*h, ") =", func(a+i*h))
    
    I += func(b)
    print("f(", b, ") =", func(b))
    
    I *= ((b-a)/(2*k))
    
    print("I:", I)
    print("Error:", 100*(abs(e_t - I)/e_t), "%")
```

```python
#Simpson's 1/3 Rule Integration Calculator

import numpy as np

def func(x):
    #return (51.69375)*np.tanh(0.18977*x)
    return x**2*np.exp(x)

#limits of integral
b = 3 
a = 0

#n is number of sections, x is the value to calculate for each segment
n = 11

#actual value for comparison
e_t = 98.42768

for k in range(1, n+1):
    print("-----------------", k, "segments--------------")
    I = func(a)
    print("f(", a, ") =", func(a))
    
    for i in range(1, k):
        h = (b-a)/k
        if(i%2 == 0):
            I += 2*(func(a+i*h))
        else:
            I += 4*(func(a+i*h))
        print("f(", a+i*h, ") =", func(a+i*h))
    
    I += func(b)
    print("f(", b, ") =", func(b))
    
    I *= (((b-a)/(k))/3)
    
    print("I:", I)
    print("Error:", 100*(abs(e_t - I)/e_t), "%")
```
